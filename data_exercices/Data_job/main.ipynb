```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "mann14/global-ai-and-data-science-job-market-20202026",
    "ai_jobs.csv"
)

print("first 5 records :", df.head())
print("data frame categories :", df.dtypes)
print("data frame columns : ", df.columns)
```

```python
df.dropna()
print("after dropping null values :", df.head())
```

```python

df_de = df[df["country"] == "Germany"]
print("Germany market")
print(df_de.head())
print("Data fram shape :")
print(df_de.shape[0])
```

```python
df_analyzed_1 = df_de.drop(columns=["company_size", "employment_type", "min_experience_years", "experience_level", "job_id", "company_type", "city", "remote_type", "industry"])


plt.style.use('_mpl-gallery')

x = df_analyzed_1['company_size']
y = df_analyzed_1['city']

fig, ax = plt.subplots()

ax.bar(x, y , width=1, linewidth=0.5)

plt.show()
```
